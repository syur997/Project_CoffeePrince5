```python
df = pd.read_excel('원본파일')
df.head()
```

```python
df = df[['결제일', '상품명', '상품별 합계']]
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df[df['상품별 합계'] != 0]
df.head()
```

```python
# 메뉴명 전처리 -> 통일
df['상품명'].replace('시그니처A', '(I)시그니처 라떼', inplace=True)
df['상품명'].replace('A시그니처', '(I)시그니처 라떼', inplace=True)

df['상품명'].replace('시그니처B ', '(H)시그니처 카푸치노', inplace=True)
df['상품명'].replace('B시그니처카푸치노', '(H)시그니처 카푸치노', inplace=True)

df['상품명'].replace('ABC아메리카노', '(I)A아메리카노', inplace=True)
df['상품명'].replace('아메리카노ABC', '(I)A아메리카노', inplace=True)

df['상품명'].replace('B아메리카노', '(I)아메리카노B', inplace=True)
df['상품명'].replace('아메리카노 _ B', '(I)아메리카노B', inplace=True)

df['상품명'].replace('B_아메리카노', '(H)아메리카노B', inplace=True)
df['상품명'].replace('아메리카노B', '(H)아메리카노B', inplace=True)

df['상품명'].replace('C아메리카노', '(I)아메리카노C', inplace=True)
df['상품명'].replace('아메리카노 _ C', '(I)아메리카노C', inplace=True)

df['상품명'].replace('C_아메리카노', '(H)아메리카노C', inplace=True)
df['상품명'].replace('아메리카노C', '(H)아메리카노C', inplace=True)
```

```python
df['결제일'] = pd.to_datetime(df['결제일'])
```

```python
df.to_csv('월별메뉴_전처리완료.csv', index=False)
```
